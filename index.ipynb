## Build a Model with Log-Transformed Features and Target

### Data Preparation

Choose up to 3 of the features you investigated, and set up an X dataframe containing the log-transformed versions of these features as well as a y series containing the log-transformed version of the target.

---

<details>
    <summary style="cursor: pointer"><b>Hint (click to reveal)</b></summary>

If you are planning log transform a feature measured in _years_ (e.g. `YearRemodAdd`) consider shifting the data first. For example, you might subtract 1900 or 1910 from the year, so that a 1% increase in year is closer to meaning 1 year rather than 20 years.

</details>

```python
# Your code here - prepare data for modeling
import statsmodels.api as sm

# Step 1: Choose and prepare the features (drop rows with zeros to safely log-transform)
features = ['GarageArea', 'GrLivArea', 'OpenPorchSF', 'SalePrice']
log_df = ames[features].copy()
log_df = log_df[(log_df > 0).all(axis=1)]  # Remove rows with zero or negative values

# Step 2: Apply log transformation to features and target
X_log = np.log(log_df[['GarageArea', 'GrLivArea', 'OpenPorchSF']])
y_log = np.log(log_df['SalePrice'])

# Step 3: Add a constant to X (for the intercept)
X_log_const = sm.add_constant(X_log)

# Step 4: Fit the OLS regression model
log_model = sm.OLS(y_log, X_log_const).fit()

# Step 5: View the model summary
print(log_model.summary())
```

### Modeling

Now build a StatsModels OLS model with a log-transformed target as well as log-transformed features.

```python
# Your code here - build a model

# Step 1: Select relevant columns and remove 0s to avoid log(0)
features = ['GarageArea', 'GrLivArea', 'OpenPorchSF', 'SalePrice']
log_data = ames[features].copy()
log_data = log_data[(log_data > 0).all(axis=1)]  # Remove rows where any value is zero or negative

# Step 2: Log-transform the features and target
X_log = np.log(log_data[['GarageArea', 'GrLivArea', 'OpenPorchSF']])
y_log = np.log(log_data['SalePrice'])

# Step 3: Add constant for intercept
X_log_const = sm.add_constant(X_log)

# Step 4: Fit the OLS model
log_model = sm.OLS(y_log, X_log_const).fit()

# Step 5: Display model summary
print(log_model.summary())
```
